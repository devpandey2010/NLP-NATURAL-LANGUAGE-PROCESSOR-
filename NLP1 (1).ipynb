**Understanding Regex(Regular Expression)**

````python
import re
text="I Love python programming"
match=re.search(r"python",text)  #using the regular expression search for the eord python in text
print(match.group())
````

````python
word="helloworld"
print(re.findall(r"l.o",word))
````

````python
log_data = """
ERROR 01-05-2023: Server crashed at 10:15 AM.
Please contact: dev123@gmail.com or rahul.kumar@company.org.
Call on 9876543210 or 987-654-3211 for support.
#urgent #serverDown
Report generated by Dev Kumar at 04/11/2003.
WARNING: server overload detected.
User: Shweta sent an email to info@domain.in
"""
````

````python
#all the emails
emails=re.findall(r"\S+@\S+",log_data)
print(emails)
````

````python
#phone numbers
phonenumber=re.findall("\d{10}|\d{3}-\d{3}-\d{4}",log_data)
print(phonenumber)
````

````python
#Replace all dates (dd-mm-yyyy or dd/mm/yyyy) with [DATE]
date=re.findall("\d{2}[-/]\d{2}[-/]\d{4}",log_data)
print(date)
# Replace all dates with [DATE]
updated_text = re.sub(r'\b\d{2}[-/]\d{2}[-/]\d{4}\b', '[Date]', log_data)

print(updated_text)
````

````python
#Extract all lines that start with "ERROR"
match=re.findall("^ERROR.*",log_data,re.MULTILINE)
'''Normally ^ matches the start of the entire string


With re.MULTILINE, ^ matches the start of each line inside a multi-line string'''

print(match)
````

````python
#Extract all hashtags
count=re.findall("\W",log_data)
cnt=0
for i in count:
  if(i=='#'):
    cnt+=1
print(cnt)
````

````python
#Count number of times "server" appears (case-insensitive)
wahh=re.findall("server|SERVER",log_data)
print(len(wahh))
````

````python
#Extract all capitalized proper names (e.g., "Dev", "Shweta", "Rahul Kumar")
name=re.findall("\b[A-Z][a-z]+(?:\s[A-Z][a-z]+)?\b",log_data)
print(name)
````

````python
text="in my name, XYZ, ABC"
match=re.findall("[^aeiou]\S+",text)
print(match)
````

````python
'''‚úÖ \S+ ‚Äî "One or More Non-Whitespace Characters"
+ means 1 or more times

So \S+ = match a full word or block of characters until a space

üîπ Example:
python
Copy code
re.findall(r"\S+", "Hi Bro! Welcome123")
# Output: ['Hi', 'Bro!', 'Welcome123']
‚û°Ô∏è Basically, it splits on whitespace and returns words

'''
text="1234 and 56"
match=re.findall("\d+\S",text)
#or can be solved by other way
rematch=re.findall("\d{1,10}",text)
print(rematch)
print(match)
````

````python
text="I like the red color,and the blue colour"
match=re.findall("colou?r",text)  #? means either 0 repeatations or 1 repeatations
print(match)
````

````python
#groups and capturing
text="Call me at 123-456-7890"
match=re.search("\d{3}-\d{3}-\d{4}",text)
rematch=re.findall("\d{3}-\d{3}-\d{4}",text)
print(rematch)
print(match.group())
````

````python
text="abc123"
match=re.search("abc(?=123)",text)  #matching abc which is precedence of 123
rematch=re.findall("abc(?=123)",text)
print(match.group())
print(rematch)
````

````python
re.findall(r"@gmail(?=\.com)", "abc@gmail.com abc@gmail.in")
# ['@gmail']
````

````python
re.findall(r"\d+(?!%)", "20% off and 500 extra")
# ['500']
````

````python
'''üß† Why \b works:
\b = word boundary, ensures the number is a full token

So partial 00 in 500 doesn‚Äôt get matched anymore'''

import re

text = "Price: ‚Çπ500 and $300"

# Match only numbers preceded by ‚Çπ
matches = re.findall(r"(?<!‚Çπ)\b\d+\b", text)
print(matches)  # ['500']
````

````python
text="Price: ‚Çπ500 and $300 and 600"
matches = re.findall(r"(?<!‚Çπ)(?<!\$)\d+", text)
print(matches)
````

````python
text="i love programming"
match=re.search(r'\bprogramming\b',text)
print(match.group())
````

````python
#Substitute
text="my phone number is 1234 and 5678"
mode=re.sub(r'\d+','4321',text)
print(mode)
````

````python
'''text = "my phone number is 1234 and 5678"
Bas 1234 ko replace karein ‚Üí '4321'
Aur 5678 waise hi rahe ‚úÖ'''

text = "my phone number is 1234 and 5678"
mode = re.sub(r'\b1234\b', '4321', text)
print(mode)
````

````python
#Escape character
text="Hello#.How are you?"
match=re.search(r"Hello#.",text)
print(match.group())
````

````python
text="there are 3 apples"
match=re.findall(r'\D+',text)
print(match)
````

````python
text="there are 3 apples"
match=re.findall(r'\S+',text)
print(match)
````

````python
#matching exact string lengths
text="abc def ghi"
match=re.findall(r'\b[a-zA-Z]{2}\b',text)
print(match)
````

````python
text="abc def ghi"
match=re.findall(r'\b[a-zA-Z]{3}\b',text)
print(match)
````

````python
#without word bounding it will also give xyz even its length is not 3 thatwhy we need to have wordbounding it will give me exact length word
text="abc def ghi xyzl"
match=re.findall(r'[a-zA-Z]{3}',text)
print(match)
````

````python
#Matching phonenumbers
text="My numbers are 123.456.7890 and (980)345-4321"
rematch=re.search(r'\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}',text)
print(rematch.group())
text="My numbers are 123.456.7890 and (980)345-4321"
match=re.findall(r'\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}',text)
print(match)
````

````python
#Matching email address
text="my email is dev123@gmail.in"
match=re.search(r'\S+\@\S+',text)  #means nonsapce+@+nonsapce
print(match.group())

rematch=re.search(r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9._%+-]+\.[a-zA-Z0-9._%+-]{2,}',text)
print(rematch.group())
````

````python
#remove whitespaces
text="Hello world: "
match=re.sub(r'\s+',".",text)
print(match)
````

````python
#Ignore the case
text="Hello match"
match=re.search(r"Hello",text,re.IGNORECASE)
print(match.group())
````

# üìå Regex (Regular Expressions) ‚Äì Summary

**Regular Expressions** (Regex) are patterns used to match, extract, or replace text. Widely used in NLP, data cleaning, validation, and text processing.

---

## üîπ Basic Symbols

| Symbol | Meaning | Example     | Matches         |
|--------|---------|-------------|-----------------|
| `\d`   | Digit (0‚Äì9)        | `\d{4}`     | `1234`        |
| `\w`   | Word character (a-z, A-Z, 0-9, _) | `\w+` | `hello_123` |
| `\s`   | Whitespace         | `\s+`       | `"   "`       |
| `.`    | Any character except newline | `a.c` | `abc`, `a-c` |
| `^`    | Start of string    | `^Hi`       | `Hi there`    |
| `$`    | End of string      | `end$`      | `the end`     |
| `*`    | 0 or more times    | `lo*l`      | `ll`, `lol`, `loool` |
| `+`    | 1 or more times    | `go+d`      | `god`, `good` |
| `?`    | 0 or 1 time        | `colou?r`   | `color`, `colour` |
| `[]`   | Match any one character inside | `[aeiou]` | `a`, `e` |
| `|`    | OR                 | `cat|dog`   | `cat` or `dog` |
| `()`   | Grouping           | `(abc)+`    | `abc`, `abcabc` |
